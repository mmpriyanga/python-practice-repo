```python
"""
Use Case 2: Student Result Classification
a. Write a program that takes marks as input (initially as a string).
B. Check if the value can be converted to float.
C. Then classify:
Marks >= 90 --> Outstanding
 Marks >= 75 --> Excellent
 Marks >= 50 --> Pass
 Marks < 50 --> Fail
D. If the input is not numeric, print:
 Invalid marks entered — Please provide numeric input.
 """

inp = input("Enter the Marks: ")

while True:
    try:
        inp = int(inp)

        # Range validation
        if inp < 0 or inp > 100:
            inp = input("Invalid entry. Please enter a number between 0 and 100: ")
            continue

        break  # valid integer and valid range → exit loop

    except ValueError:
        inp = input("Invalid entry. Please enter a numeric value: ")

# Classification
if inp >= 90:
    print("Outstanding")
elif inp >= 75:
    print("Excellent")
elif inp >= 50:
    print("Pass")
else:
    print("Fail")
```

```python
while True:
    try:
        total_monthly_data = float(input("Enter your Total monthly data limit (in GB): "))
        data_used = float(input("Enter data used so far (in GB): "))

        # Value validation
        if total_monthly_data <= 0:
            print("Monthly limit must be greater than 0. Try again.\n")
            continue

        if data_used < 0:
            print("Used data cannot be negative. Try again.\n")
            continue

        if data_used > total_monthly_data:
            print("Used data cannot be greater than the limit. Give valid input again.\n")
            continue

        # Valid input → break loop
        remaining_data = total_monthly_data - data_used
        usage_perc = (data_used / total_monthly_data) * 100
        break

    except ValueError:
        print("Invalid entry. Please enter a numeric value.\n")

# Output section
print("\nThe Remaining data is:", remaining_data)

if usage_perc >= 80:
    print("Warning: High usage, consider upgrading your plan.")
else:
    print(f"The usage percent is {round(usage_perc, 2)}")
```
